```python
#Quality of life changes
    #renaming columns to '_'
    #renaming columns to lowercase
df.rename(columns={"SeniorCitizen":"senior_citizen",
                  "PhoneService":"phone_service",
                  "MultipleLines":"multiple_lines",
                  "InternetService":"internet_service",
                  "OnlineSecurity":"online_security",
                  "OnlineBackup":"online_backup",
                  "DeviceProtection":"device_protection",
                  "TechSupport":"tech_support",
                  "StreamingTV":"streaming_tv",
                  "StreamingMovies":"streaming_movies",
                  "PaperlessBilling":"paperless_billing",
                  "PaymentMethod":"payment_method",
                  "MonthlyCharges":"monthly_charges",
                  "TotalCharges":"total_charges"},inplace=True)
df.columns = df.columns.str.lower()
```

```python
# Data qualiity checks
    # checking for duplicated values 
    # checking for missing data
assert df.duplicated().all() == False, "not today mate"
#try:
#    df.isna().any()
```

```python
#checking for missing data 
#remember that missing data can be represented in different ways, for example 'Nan', and unusual values 
# like -99, -100.
try:
    df.isna().any()[1] == False
except:
    print("That there missing data with the value of NaN")
df.describe(include='all').T
```

# Changing datatypes 

```python
# unusual for column 'total_charges' to be classed as an object when it should be at least numerical
```

```python
df.info()
```

```python
# due to it being a string, most likley these values are recorded as ' ' 
# If true, the assumption is that this is due them being new customers 
assert((df.total_charges == ' ').any()), "not today mate"
assert(((df.total_charges == ' ') == (df.tenure == 0)).all())
df.total_charges = df.total_charges.replace({' ': '0'})
```

```python
# changing the column total_charges to a numerical datatype
df.total_charges = pd.to_numeric(df.total_charges)
#df.total_charges.as_type(float)
```
